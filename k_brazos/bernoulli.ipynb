## Experimento

En este notebook se aborda el problema del bandido multi-brazo para un entorno de 10 brazos, donde la recompensa de cada alternativa modela escenarios de éxito o fracaso regidos por una distribución de Bernoulli. El objetivo de este estudio es analizar y comparar la capacidad de distintas familias de algoritmos para encontrar el equilibrio óptimo entre la exploración del entorno y la explotación del conocimiento adquirido.

Para llevar a cabo un análisis riguroso, el experimento se estructura en tres capas de evaluación:

1. **Comparativa de parámetros:** Se analiza qué combinación de parámetros arroja los mejores resultados para un algoritmo determinado, permitiendo así comprender y medir la influencia que tienen estos valores sobre la capacidad de aprendizaje del agente.

2. **Selección interna:** Se evalúa el rendimiento dentro de una misma familia de algoritmos para identificar la mejor variante dentro de cada enfoque. Por un lado, se comparan las políticas epsilon-greedy y epsilon-decaimiento; por otro, se enfrentan los algoritmos UCB1 y UCB2.

3. **Comparativa final:** Se realiza un enfrentamiento cruzado entre el algoritmo Softmax y los dos algoritmos que hayan demostrado un desempeño superior en la capa anterior (el mejor de la familia epsilon y el mejor de la familia UCB).

Cada política se evalúa a lo largo de 1000 pasos, repitiendo el experimento completo 500 veces de forma independiente y utilizando una semilla de generación de números aleatorios fija para garantizar la total reproducibilidad de los resultados. En cada iteración, se calcula el promedio de las métricas obtenidas sobre estas 500 ejecuciones.

La diferentes comparativas se fundamentan en el análisis de las métricas exigidas para este tipo de problemas: la evolución de la recompensa media, el porcentaje en el que fue seleccionado el brazo óptimo y la evolución del rechazo (regret) acumulado a lo largo del tiempo.
